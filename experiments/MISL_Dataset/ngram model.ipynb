```python
import os, sys
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

### Results for old dataset (hand accumulated)

```python
experiment_file = "ngram.json"
truth_file = "HLA-A*02:01-true.json"
predict_file = "HLA-A*02:01-predict.json"
with open(experiment_file, 'r') as f:
    data = json.load(f)

with open(truth_file, 'r') as f:
    scatter_truth = json.load(f)

with open(predict_file, 'r') as f:
    scatter_predict = json.load(f)
```

```python
# Visualizations for a Lasso Linear Regression Model
varying_k = list(data.keys())

hundred_features = []
two_hundred_features = []
three_hundred_features = []
four_hundred_features = []
five_hundred_features = []
six_hundred_features = []
seven_hundred_features = []
eight_hundred_features = []
nine_hundred_features = []
thousand_features = []

for k in data.keys():
    vals = data[k]
    for v in vals:
        if v[0] == 100:
            hundred_features.append(v[2])
        elif v[0] == 200:
            two_hundred_features.append(v[2])
        elif v[0] == 300:
            three_hundred_features.append(v[2])
        elif v[0] == 400:
            four_hundred_features.append(v[2])
        elif v[0] == 500:
            five_hundred_features.append(v[2])
        elif v[0] == 600:
            six_hundred_features.append(v[2])
        elif v[0] == 700:
            seven_hundred_features.append(v[2])
        elif v[0] == 800:
            eight_hundred_features.append(v[2])
        elif v[0] == 900:
            nine_hundred_features.append(v[2])
        else:
            thousand_features.append(v[2])

plt.title("Test Set MSE Loss across different ngram hyperparameters")
plt.xlabel("K value")
plt.ylabel("Sum of MSE across all alleles")
plt.plot(varying_k, hundred_features, label="100 features")
plt.plot(varying_k, two_hundred_features, label="200 features")
plt.plot(varying_k, three_hundred_features, label="300 features")
plt.plot(varying_k, four_hundred_features, label="400 features")
plt.plot(varying_k, five_hundred_features, label="500 features")
plt.plot(varying_k, six_hundred_features, label="600 features")
plt.plot(varying_k, seven_hundred_features, label="700 features")
plt.plot(varying_k, eight_hundred_features, label="800 features")
plt.plot(varying_k, nine_hundred_features, label="900 features")
plt.plot(varying_k, thousand_features, label="1000 features")
plt.legend()
plt.show()
```

### MHC Flurry Dataset

```python
regression_file = "ngram.json"
classification_file = "ngram_svm.json"
with open(regression_file, 'r') as f:
    regression_data = json.load(f)
    
with open(classification_file, 'r') as f:
    classification_data = json.load(f)
```

```python
def visualize(data, title, ylabel):
    varying_k = list(data.keys())
    
    print(str(varying_k))

    hundred_features = []
    two_hundred_features = []
    three_hundred_features = []
    four_hundred_features = []
    five_hundred_features = []
    six_hundred_features = []
    seven_hundred_features = []
    eight_hundred_features = []
    nine_hundred_features = []
    thousand_features = []
    
    for k in data.keys():
        vals = data[k]
        for v in vals:
            if v[0] == 100:
                hundred_features.append(v[2])
            elif v[0] == 200:
                two_hundred_features.append(v[2])
            elif v[0] == 300:
                three_hundred_features.append(v[2])
            elif v[0] == 400:
                four_hundred_features.append(v[2])
            elif v[0] == 500:
                five_hundred_features.append(v[2])
            elif v[0] == 600:
                six_hundred_features.append(v[2])
            elif v[0] == 700:
                seven_hundred_features.append(v[2])
            elif v[0] == 800:
                eight_hundred_features.append(v[2])
            elif v[0] == 900:
                nine_hundred_features.append(v[2])
            else:
                thousand_features.append(v[2])

    plt.title(title)
    plt.xlabel("K value")
    plt.ylabel(ylabel)

    plt.plot(varying_k, hundred_features, label="100 features")
    plt.plot(varying_k, two_hundred_features, label="200 features")
    plt.plot(varying_k, three_hundred_features, label="300 features")
    plt.plot(varying_k, four_hundred_features, label="400 features")
    plt.plot(varying_k, five_hundred_features, label="500 features")
    plt.plot(varying_k, six_hundred_features, label="600 features")
    plt.plot(varying_k, seven_hundred_features, label="700 features")
    plt.plot(varying_k, eight_hundred_features, label="800 features")
    plt.plot(varying_k, nine_hundred_features, label="900 features")
    plt.plot(varying_k, thousand_features, label="1000 features")
    plt.legend()
    plt.show()
```

```python
visualize(regression_data, "Sum of Test Set MSE", "MSE")
```

```python
visualize(classification_data, "Average classification accuracy across alleles", "Accuracy")
```
