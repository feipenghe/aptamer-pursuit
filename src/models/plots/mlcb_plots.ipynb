# MLCB Paper Plots and Figures

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

## Blosum Clipped Encoding

```python
# Experimental Replicate AUCs
plt.figure(num=None, figsize=(5, 5))
test_auc_1 = {'LinearNet': 0.85, 'ConvNet': 0.7865, 'Resnet': 0.671}
test_auc_2 = {'LinearNet': 0.8605, 'ConvNet': 0.541, 'Resnet': 0.67}
test_auc_3 = {'LinearNet': 0.866, 'ConvNet': 0.533, 'Resnet': 0.6905}

plt.scatter(test_auc_1.keys(), test_auc_1.values(), s=200, marker='>', label='Test Set 1')
plt.scatter(test_auc_2.keys(), test_auc_2.values(), s=200, marker='*', label='Test Set 2')
plt.scatter(test_auc_3.keys(), test_auc_3.values(), s=200, marker='p', label='Test Set 3')

plt.legend()
```
