```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# ROC Curves for deciding the best Lambda_val

## BlosumLinearNet

```python
lambda_vals = [1, 1.5, 2, 2.5, 2.75, 3, 4]
tpr = [74.6, 76.5, 76.2, 73.6, 74.5, 75, 72.7]
tnr = [73.5, 72.19, 70.1, 73.7, 73.36, 72.6, 75.5]
fpr = [100-x for x in tnr]
print(str(fpr))
```

```python
plt.title("ROC Curve BlosumLinearNet")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
sns.lineplot(fpr, tpr)
```
